```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
f = lambda x: pow(x,2)

xs = [2,3,4]

xs_bin = []

for x in xs:
    binary = format(x,'b')
    xs_bin.append(binary)

    

f_x = [f(x) for x in xs]

f_prop = []
for x in f_x:
    prop = x / sum(f_x)
    f_prop.append(prop)

print('x values 1: ', xs)
print('binary x values 1:', xs_bin)
print('fitness 0: ', f_x)
print('fitness proporionalities 0: ', f_prop)


fig, ax = plt.subplots()
ax.pie(f_prop, labels=xs, autopct='%.1f%%')
ax.set_aspect('equal') 

# We can see here that x=4 achieves the highest fitness for the function f(x). 
# Therefore it also obtains the highest fitness proporionality of 0.55
```

```python
# Same code as above, however we change the function of f(x)
f = lambda x: pow(x,2) + 20

xs = [2,3,4]

xs_bin = []

for x in xs:
    binary = format(x,'b')
    xs_bin.append(binary)

    

f_x = [f(x) for x in xs]

f_prop = []
for x in f_x:
    prop = x / sum(f_x)
    f_prop.append(prop)

print('x values 0: ', xs)
print('binary x values 0:', xs_bin)
print('fitness 0: ', f_x)
print('fitness proporionalities 0: ', f_prop)


fig, ax = plt.subplots()
ax.pie(f_prop, labels=xs, autopct='%.1f%%')
ax.set_aspect('equal') 

# We can see here that x=4 achieves the highest fitness for the function f(x). 
# Therefore it also obtains the highest (albeit more equal) fitness proporionality of 0.40
```

```python
l = 100
p = 1/l
fitness = lambda x: sum(x)
iterations = 1500
x = [random.randint(0, 1) for _ in range(l)]
goal_runs = 1
iterations = 1500


runs_lst = []

runs = 0
while runs < goal_runs:
    iteration = 0
    iteration_lst = []
    while len(iteration_lst) < iterations:
        x_m = x.copy()
        for index in range(len(x_m)):
            rand = random.uniform(0, 1)
            if rand <= p:
                if x_m[index] == 1:
                    x_m[index] = 0
                else:
                    x_m[index] = 1        
        
        mutate_score = fitness(x_m)
        origin_score = fitness(x)

        if mutate_score > origin_score:
            x = x_m
        iteration_lst.append(origin_score)
    runs_lst.append(iteration_lst)
    runs += 1
    

plt.title('10 runs with random starting values for x')
for index, run in enumerate(runs_lst):
    plt.plot(run, label = 'Run {0:2d}'.format(index))
plt.ylim([40,101])
plt.xlabel('Iteration')
plt.ylabel('Fitness')
```

```python
l = 100
p = 1/l
fitness = lambda x: sum(x)
iterations = 1500
x = [random.randint(0, 1) for _ in range(l)]
goal_runs = 10
iterations = 1500


runs_lst = []

runs = 0
while runs < goal_runs:
    iteration = 0
    best_score = 0
    iteration_lst = []
    x = [random.randint(0, 1) for _ in range(l)]
    while len(iteration_lst) < iterations:
        x_m = x.copy()
        for index in range(len(x_m)):
            rand = random.uniform(0, 1)
            if rand <= p:
                if x_m[index] == 1:
                    x_m[index] = 0
                else:
                    x_m[index] = 1        
        
        mutate_score = fitness(x_m)

        if mutate_score > best_score:
            best_score = mutate_score
        iteration_lst.append(best_score)
    runs_lst.append(iteration_lst)
    runs += 1
    

plt.title('10 runs with random starting values for x')
for index, run in enumerate(runs_lst):
    plt.plot(run, label = 'Run {0:2d}'.format(index))
plt.ylim([40,101])
plt.xlabel('Iteration')
plt.ylabel('Fitness')
```

10 runs with 1500 iterations each, where each mutation is saved and used as the basis for the next iteration. This means the best mutation is not saved, therefore the algorithm will probably never converge to 100 fitness. The plot depitcts the best achieved scores per run.

```python
def load_file(name):
    tsp = []
    with open(name) as f:
        lines = f.readlines()
        
    for line in lines:
        tsp_single = []
        coordinates = line.split()
        tsp_single.append(float(coordinates[0]))
        tsp_single.append(float(coordinates[1]))
        tsp.append(tsp_single)

    return tsp

def distance(coordinate_1, coordinate_2):
    x = coordinate_1[0] - coordinate_2[0]
    y = coordinate_1[1] - coordinate_2[1]
    distance = np.sqrt(x**2 + y**2)
    return distance


def total_distance(route):
    total_dist = 0
    for index, city in enumerate(route):
        if index < len(route)-1:
            next_city = route[index+1]
            total_dist += distance(city, next_city)
    return total_dist

def local_search(list):
    new_list = []
    pos1 = random.randint(0,len(list)-2)
    pos2 = random.randint(pos1+1,len(list)-1)
    
    for i in range(0,pos1):
        new_list.append(list[i])
    
    for j in reversed(range(pos1,pos2+1)):
        new_list.append(list[j])
    
    for k in range(pos2+1, len(list)):
        new_list.append(list[k])
        
#     if total_distance(new_list) < total_distance(list):
#         return list
    
    return new_list


def create_pool(candidates, tsp):
    pool = []
    for i in range(candidates):
        pool_candidate = np.random.permutation(tsp)
        pool.append(pool_candidate.tolist())
    return pool

def crossover(p1, p2): 
    cut_1 = random.randint(1,len(p1)-2)
    cut_2 = random.randint(cut_1+1,len(p2)-1)
    
    p1_inside = p1[cut_1:cut_2]
    p1_outside = p1[cut_2:] + p1[:cut_2]
    
    p2_inside = p2[cut_1:cut_2]
    p2_outside = p2[cut_2:] + p2[:cut_2]
    
    p1_rest = [x for x in p1_outside if x not in p2_inside]
    p2_rest = [x for x in p2_outside if x not in p1_inside]
    
    o1 = p2_rest[len(p2)-cut_2:] + p1_inside + p2_rest[:len(p2)-cut_2]
    o2 = p1_rest[len(p1)-cut_2:] + p2_inside + p1_rest[:len(p1)-cut_2]
    
    return o1, o2


def mutate(list):
    i, j = np.random.choice(range(len(list)), 2, replace = False) 
    list[i], list[j] = list[j], list[i]
    return list


def create_next_generation(pool, prop_fitnesses):
    parents = random.choices(pool, prop_fitnesses, k = len(pool))
    
    new_generation = []
    loop_max = int(np.ceil(len(parents)/2))
    for i in range(loop_max):
        p1, p2 = random.sample(parents, 2)        
        o1, o2 = crossover(p1, p2)
        
        o1 = mutate(o1)
        o2 = mutate(o2)
        
        o1 = local_search(o1)
        o2 = local_search(o2)
        
        new_generation.append(o1)
        new_generation.append(o2)
        
    # Drop final individual when pool size is odd?
    if not len(pool)%2 == 0:
        index = random.randint(0,len(pool))
        del new_generation[index]
        
    return new_generation
        
def two_opt_algorithm(tsp, pool_size, max_iterations):
    pool = create_pool(pool_size, tsp)
    for iteration in range(max_iterations):
        fitnesses = [1/total_distance(route) for route in pool]
        prop_fitnesses = [route/sum(fitnesses) for route in fitnesses]

        pool = create_next_generation(pool, prop_fitnesses)
        print('avg fitness: ', sum(fitnesses) / len(fitnesses))

        
        
tsp = load_file('file-tsp.txt')
pool_size = 16
runs = 100

new_tsp = two_opt_algorithm(tsp, pool_size, runs)

```
